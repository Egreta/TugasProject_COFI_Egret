```python
import pandas as pd

df=pd.read_csv('USairpollution_quiz2.csv')
df
```

```python
#your journey & code Start here, Good luck!!!
df_sort=df.sort_values(by='clusters')
df_sort
```

```python
clasters_precip=df_sort.groupby('clusters')['precip'].agg(['max','min','mean']).reset_index()
clasters_precip
```

```python
clasters_precip['mean']=clasters_precip['mean'].astype(int)

plt.bar(clasters_precip['clusters'],clasters_precip['mean'])
#plt.xticks(rotation=65)
plt.title('Rata-rata Curah Hujan')
plt.xlabel('claster')
plt.ylabel('inches')
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
        
addlabels(clasters_precip['clusters'], clasters_precip['mean'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = df_sort['precip']
y = df_sort['predays']
z = df_sort['clusters']

fig = plt.figure(figsize=(6, 6))
plt.scatter(x, y,
           linewidths=1, alpha=.7,
           edgecolor='k',
           s = 200,
           c=z)
plt.xlabel('Precip')
plt.ylabel('Predays')
plt.title('Comparing Clusters with the Precis and Predays feature')
plt.show()
```

```python
clasters_predays=df_sort.groupby('clusters')['predays'].agg(['min','max','mean']).reset_index()
clasters_predays
```
